
# Test if your newly deployed endpoints work....

In order to get this to work, please ensure that you have configured the aws credentials to have permissions to invoke endpoint. For more details, see  https://boto3.amazonaws.com/v1/documentation/api/latest/guide/configuration.html.


## Configure the endpoint name and region

```python
endpointname="DummyS2"
region="ap-southeast-2"
```

```python
import boto3
import json
```

```python
data={
    "instances": [
        {
            "start": "2009-11-01 00:00:00",
            "target": [4.0, 10.0, "NaN", 100.0, 113.0]
           
        }
    ],
    "configuration": {
         "num_samples": 50,
         "output_types": ["mean", "quantiles", "samples"],
         "quantiles": ["0.5", "0.9"]
    }
}
```

```python

client = boto3.client('sagemaker-runtime', region_name=region)
```

```python
response = client.invoke_endpoint(
    EndpointName=endpointname,
    Body= json.dumps(data).encode(),
    ContentType='application/json',
    Accept='application/json'
)

body = response["Body"].read()
```

```python
result = json.loads(body.decode())
```

```python
print(json.dumps(result, indent=2))
```
